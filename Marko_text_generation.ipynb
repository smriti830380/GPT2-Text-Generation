```python
import random

text = """Artificial intelligence is the future.
AI is powerful and useful.
Machine learning is a part of AI."""

words = text.split()

# Markov Chain dictionary
markov_dict = {}

for i in range(len(words)-1):
    word = words[i]
    next_word = words[i+1]

    if word not in markov_dict:
        markov_dict[word] = []

    markov_dict[word].append(next_word)

# Generate text
current_word = random.choice(words)
generated_text = current_word

for _ in range(20):
    if current_word in markov_dict:
        next_word = random.choice(markov_dict[current_word])
        generated_text += " " + next_word
        current_word = next_word
    else:
        break

print(generated_text)
```
